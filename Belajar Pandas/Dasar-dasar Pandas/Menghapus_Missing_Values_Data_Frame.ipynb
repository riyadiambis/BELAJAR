```python
import pandas as pd
import numpy as np

print(pd.__version__)
```

Membuat data yang memiliki missing di dalamnya

Ada dua pendekatan yakni dengan menghapuys kolom yang mengandung missing values atau menghapus baris yang mengandung missing.

```python


# Membuat DataFrame dengan beberapa nilai yang hilang
df = pd.DataFrame({
    'A': [1, 2, np.nan, 4],
    'B': [np.nan, 2, 3, 4],
    'C': [1, np.nan, np.nan, 4],
    'D': [2, 4, 5, 1090]
})

# Mereset index
df = df.reset_index(drop=True)

# Menampilkan 5 baris pertama DataFrame
print(df.head())
```

```python
#Mngubah nama kolom A menjadi WIBU

df = df.rename(columns={"A":'Wibu'})
df
```

```python
df_backup = df.copy(deep=True)
```

```python
#Ini adalah untuk menghapus kolom yang memiliki nilai NaN
df = df.dropna(axis='columns')
df.head()
```

Kita akan mengembalikan nilai df seperti semula dengan memanfaatkan hasil copy kita sebelumnya

```python
df = df_backup.copy(deep=True)

#Kita akan mencoba melakukan pembersihan kolom

df = df.dropna(axis='rows')
df.head()
```

isna() di sini akan mengumpulkan setiap missing valueas yang ada di dalam dataframe

mean() akan mengkalkulasi rata-rata dari missing values dari data frame

Tujuan perhitungan ini memungkinkan kita mengisi missing values dengan nilai mean apabila persentase kolom itu masih bagus

```python
df = df_backup.copy(deep=True)
df.isna().mean()
```

Kita akan mencoba membatasi kolom yang pantas kita hapus atau tidak, kita akan menggunakan thresh dan menentukan persentasenya apakah pantas atau enggak untuk di hapus. nilai thresh akan memuat ambang batas yang telah kita tetapkan dan akan menjadi syarat kelayakan penghapusan.

```python
treshhold = len(df)*0.7
df = df. dropna(thresh=treshhold, axis='columns')
df
```
